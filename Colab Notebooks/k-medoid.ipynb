```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn_extra.cluster import KMedoids

# Load the Iris dataset
iris = load_iris()
X = iris.data

# K-Medoids clustering
kmedoids = KMedoids(n_clusters=3, random_state=42)
y_kmedoids = kmedoids.fit_predict(X)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmedoids, s=50, cmap='viridis', label='Data Points')

# Get the coordinates of the medoids
medoid_indices = kmedoids.medoid_indices_
plt.scatter(X[medoid_indices, 0], X[medoid_indices, 1],
            c='red', s=200, alpha=0.5, marker='X', label='Medoids')

plt.title('K-Medoids Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.grid()
plt.show()
```
