```python
import pandas as pd
import numpy as np

# Load the Iris dataset from a URL
dataset_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# Define the column names
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']

# Load the dataset into a DataFrame
iris = pd.read_csv(dataset_url, names=columns)

# Display the first few rows of the dataset to understand its structure
print("First few rows of the dataset:")
print(iris.head(), "\n")

# Calculate the mean of each numeric column
mean_values = iris.mean(numeric_only=True)
print("Mean of each numeric column:")
print(mean_values, "\n")

# Calculate the median of each numeric column
median_values = iris.median(numeric_only=True)
print("Median of each numeric column:")
print(median_values, "\n")

# Calculate the mode of each numeric column
mode_values = iris.mode().iloc[0]  # Taking the first mode value if there are multiple modes
print("Mode of each numeric column:")
print(mode_values, "\n")

# Generate summary statistics for the dataset
summary_statistics = iris.describe()
print("Summary statistics of the dataset:")
print(summary_statistics)
```
