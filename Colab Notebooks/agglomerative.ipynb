```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

# Sample data
x = [[i] for i in [12, 10, 6, 8, 4, 2, 1, 16]]

# Perform hierarchical clustering using the Ward method
Z_ward = linkage(x, method='ward')

# Create a dendrogram for Ward's method
plt.figure(figsize=(25, 10))
dendrogram(Z_ward)
plt.title('Dendrogram (Ward\'s Method)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.grid()
plt.show()

# Perform hierarchical clustering using the Single linkage method
Z_single = linkage(x, method='single')

# Create a dendrogram for Single linkage
plt.figure(figsize=(25, 10))
dendrogram(Z_single)
plt.title('Dendrogram (Single Linkage Method)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.grid()
plt.show()


```
