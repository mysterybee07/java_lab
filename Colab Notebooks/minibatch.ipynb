```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import MiniBatchKMeans

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Mini-Batch K-Means clustering
minibatch_kmeans = MiniBatchKMeans(n_clusters=3, random_state=42)
y_minibatch_kmeans = minibatch_kmeans.fit_predict(X)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_minibatch_kmeans, s=50, cmap='viridis', label='Data Points')
plt.scatter(minibatch_kmeans.cluster_centers_[:, 0], minibatch_kmeans.cluster_centers_[:, 1],
            c='red', s=200, alpha=0.5, marker='X', label='Centroids')
plt.title('Mini-Batch K-Means Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.grid()
plt.show()
```
